```python
import yfinance as yf
import mplfinance as mpf
from datetime import datetime as dtdt
```

```python
ticker = 'BTC-USD'
year=2021
month=1
day=15
volume = 'yes'
moving_average = 20
```

```python
print(f'Selected ticker: {ticker}')
```

```python
date = dtdt.strptime(f'{year}-{month}-{day}', '%Y-%m-%d')
barras = (dtdt.today() - date).days
tipo = 'candle' if barras <= 598 else 'line'
Volume = True if volume == 'yes' else False
```

```python
data = yf.download(ticker, start=date, auto_adjust=True, progress=False)
```

```python
print(f'Current price: {"${:,.2f}".format(data.Close[-1])}')
print(f'Period\'s low: {"${:,.2f}".format(data.Low.min())}')
print(f'Period\'s high: {"${:,.2f}".format(data.High.max())}\n\n')
```

```python
print('Last five trading days:\n')
data.tail(5)
```

```python
mpf.plot(data, type=tipo, style='charles', mav=(moving_average), volume=Volume, figratio=(25,10))
```
