```python
import yfinance as yf
import mplfinance as mpf
```

```python
ticker = 'GGAL'
```

```python
print(f'Selected ticker: {ticker}')
```

```python
data = yf.download(ticker, period='ytd', auto_adjust=True)
```

```python
print(f'YTD low: {round(data.Low.min(),2)}')
print(f'YTD high: {round(data.High.max(),2)}')
```

```python
data.tail()
```

```python
mpf.plot(data, type='candle', style='charles', mav=(20),figratio=(18,10))
```
